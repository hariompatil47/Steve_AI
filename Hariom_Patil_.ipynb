#1. Given an integer n representing the number of pairs of parentheses, write a function to generate all
Combinations of well-formed parentheses.

*Example* :

***Input** :-

An integer **n (1 ≤ n ≤ 8)**, where n is the number of pairs of parentheses.


**Output** :-

A list of strings, where each string represents a valid combination of well-formed parentheses.

The combinations should be returned in any order.


*Example*

**Input**:-

n = 3

**Output** :-

[&quot;((()))&quot;, &quot;(()())&quot;, &quot;(())()&quot;, &quot;()(())&quot;, &quot;()()()&quot;]

**Input** :-

n= 1

**Output** :-
[&quot;()&quot;]

````python
# Write here solution
def generate(n):
    result = []

    def backtrack(current, open_count, close_count):
        if len(current) == 2 * n:
            result.append(current)
            return
        if open_count < n:
            backtrack(current + '(', open_count + 1, close_count)

        if close_count < open_count:
            backtrack(current + ')', open_count, close_count + 1)

    backtrack('', 0, 0)
    return result
````

# 3. A subsequence of a string is a sequence derived from the original string by deleting zero or more characters without changing the order of the remaining characters.
Write a recursive function in Python to generate all possible subsequences of a given string.
The function should take the input string and return a list of all its subsequences, including the empty subsequence.

*Example*

**Input**:

"abc"

**Output**:

["", "a", "b", "c", "ab", "ac", "bc", "abc"]

````python
# Write here solution
def longest_common(text1, text2):
    m, n = len(text1), len(text2)

    prev = [0] * (n + 1)
    curr = [0] * (n + 1)

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                curr[j] = prev[j - 1] + 1
            else:
                curr[j] = max(prev[j], curr[j - 1])
        prev, curr = curr, prev

    return prev[m]
````

# 4. You are given a list of integers and a target sum. Your task is to write a function to find all unique pairs of numbers in the list that add up to the target sum.

A pair of numbers is considered unique if the numbers are not repeated in the pair or in the output. The order of the numbers in the pair does not matter, and a number cannot be paired with itself unless it appears multiple times in the list.

Example

**Input**:

arr = [2, 4, 3, 5, 7, 8]

target = 9


**Output**:

[(2, 7), (4, 5)]



````python
# Write here solution
def findpairswith(nums, target):
    nums_sorted = sorted(nums)
    pairs = []
    left, right = 0, len(nums_sorted) - 1

    while left < right:
        current_sum = nums_sorted[left] + nums_sorted[right]

        if current_sum == target:
            pair = [nums_sorted[left], nums_sorted[right]]
            pairs.append(pair)

        elif current_sum < target:
            left += 1
        else:
            right -= 1
    return pairs
````
